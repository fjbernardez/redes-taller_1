```python
S1 = {}
def probabilidad(apariciones,total):
  return apariciones/total
def informacion(apariciones,total):
  return math.log2(1/probabilidad(apariciones,total))
def mostrar_fuente(S):
    N = sum(S.values())
    simbolos = sorted(S.items(), key=lambda x: -x[1])
    print("\n".join([ " %s : Probabilidad %.5f : Información %.5f " % (d,probabilidad(k,N),informacion(k,N)) for d,k in simbolos ]))
    entropia = 0
    for d,k in simbolos:
      entropia += probabilidad(k,N)* math.log2(1/probabilidad(k,N))
    print("Entropia de la fuente %.5f con %i paquetes leidos \n " % (entropia,N))
def callback(pkt):
    if pkt.haslayer(Ether):
        dire = "BROADCAST" if pkt[Ether].dst=="ff:ff:ff:ff:ff:ff" else "UNICAST"
        proto = pkt[Ether].type # El campo type del frame tiene el protocolo
        s_i = (dire, proto) # Aca se define el simbolo de la fuente
        if s_i not in S1:
            S1[s_i] = 0.0
        S1[s_i] += 1.0
    mostrar_fuente(S1)

```

```python
scapy_cap = rdpcap('completedCapture.pcap')
for packet in scapy_cap:
    callback(packet)
```

```python
import matplotlib.pyplot as plt
simbolos = sorted(S1.items(), key=lambda x: -x[1])
countBroad = 0
countUni = 0
for d,k in simbolos:    
    if d[0] == 'BROADCAST':
        countBroad += k        
    else:
        countUni += k
explode = (0, 0.5)

plt.pie([countBroad,countUni], explode=explode, labels=['BROADCAST','UNICAST'], autopct='%1.1f%%')
plt.savefig("piechartR1UvsB.png")
```

```python
protocols = []
K = []
percent=[]
for p,k in simbolos:
    protocols.append((p))
    K.append(k)
    percent.append(str((k)/(100*len(simbolos))) + '%')
explode=(0,0.1,0.2,0.3,0.4)

patches, texts = plt.pie(K,explode=explode)

zipped= zip(protocols,percent)
plt.legend(patches, zipped, loc="best")
plt.savefig("piechartR1Protocols.png")
```

```python
import socket
def proto_name_by_num(proto_num):
    for name,num in vars(socket).items():
        if name.startswith("IPPROTO") and proto_num == num:
            return name[8:]
    return "Protocol not found"
```

```python
protocols = []
K = []
percent=[]
for p,k in simbolos:
    protocols.append((p[0],proto_name_by_num(p[1])))
    K.append(k)
    percent.append(str((k)/(100*len(simbolos))) + '%')
explode=(0,0.1,0.2,0.3,0.4)

patches, texts = plt.pie(K,explode=explode)

zipped= zip(protocols,percent)
plt.legend(patches, zipped, loc="best")
```
